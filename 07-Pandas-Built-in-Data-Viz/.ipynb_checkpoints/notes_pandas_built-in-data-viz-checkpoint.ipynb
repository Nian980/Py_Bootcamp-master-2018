## Side Note: Style Sheets

Matplotlib has [style sheets](http://matplotlib.org/gallery.html#style_sheets) you can use to make your plots look a little nicer. These style sheets include **plot_bmh,plot_fivethirtyeight,plot_ggplot** and more. They basically create a set of style rules that your plots follow. I recommend using them, they make all your plots have the same look and feel more professional. You can even create your own if you want your company's plots to all have the same look (it is a bit tedious to create on though).

Here is how to use them.

```python
import numpy as np
import pandas as pd
import seaborn as sns #we can then set seaborn styles and pandas graphs will look like them
%matplotlib inline
```

```python
df1 = pd.read_csv('df1',index_col=0) #index is first column (a time series)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2
```

```python
df1['A'].hist(bins=20)

#notice how it's calling matplotlib under the hood
```

```python
df1['A'].plot(kind='hist',bins=20)
```

```python
df1['A'].plot.hist() #we'll be using this method
```

```python
df2.plot.area(alpha=0.5)

#This is basically the same thing as a line plot, but the area below it has been filled in with some color
# Note that area plots are stacked by default.
```

```python
df2.plot.area(alpha=0.5, stacked=False)

#unstacked area plot
```

```python
df2.plot.bar(stacked=True) #stacks the bars on top of each other
#note that it used the index as categories for the bars. We only have 9 indices so this works fine. 
```

```python
df1.plot.line(y = 'B', figsize=(12,3), lw=0.5) #lw - linewidth

#we can add other things from matplotlib like above
```

```python
df1['B'].plot() #this way works too! lineplots are the pandas default
```

```python
df1.plot.scatter(x='A', y='B',c='C', cmap='coolwarm') #c = color

#c gives it a 3rd dimension of information, the depth of the color of the dot is the C value of that (A,B) point
```

```python
df1.plot.scatter(x='A', y='B',s=df1['C']*100) #s = size, note we need to pass in df column here

#We can show the C column info by size of the dot instead of color.
#*100 factor to make the dot larger
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.head()
```

```python
#plot bivariate data (2 variables), like above
df.plot.hexbin(x='a',y='b',gridsize=25) #gridsize can enlarge the hex bins size
#can change cmap here
```

```python
#kde plot:
df2['a'].plot.kde()

#same thing if we call df2['a'].plot.density()
```

```python
df2.plot.density()
```
