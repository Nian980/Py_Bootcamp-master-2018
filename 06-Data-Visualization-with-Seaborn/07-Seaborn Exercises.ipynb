# Exercises

** Recreate the plots below using the titanic dataframe. There are very few hints since most of the plots can be done with just one or two lines of code and a hint would basically give away the solution. Keep careful attention to the x and y labels for hints.**

** *Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!* **

```python
# CODE HERE
jp = sns.jointplot(x='fare',y='age',data=titanic, xlim=(-100,600)) #For joinplot, set xlim ylim inside
jp.ax_joint.set_yticks([0,20,40,60,80]) #this is changing the tick numbers for y-axis. 
#  Note this is different from set_yscale!!!! Scale is e.g. linear, log, exponential...etc.

#jp.ax_joint gets the axes of the jointplot 
```

```python
# CODE HERE
dp = sns.distplot(titanic['fare'],kde=False,bins=30,color='red') 
dp.set_xlim(1,600) #for distplots, set xlim and ylim OUTSIDE here
```

```python
# CODE HERE
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
# CODE HERE
sp = sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
sp.set_yticks([0,20,40,60,80])
```

```python
# CODE HERE
sns.countplot(x='sex',data=titanic)
```

```python
# CODE HERE
#We need to make data into matrix form first:
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
# CODE HERE
g = sns.FacetGrid(data=titanic,col='sex',xlim=(0,80)) #creates the empty grid

g.map(plt.hist,'age')

#I thought it was: g.map(sns.distplot,'age',kde=False,bins=10) ... it's similar though
```
