# Style and Color:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# set_style sets the style for the rest of your plots

sns.set_style('whitegrid') #there's also 'darkgrid'

sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')

sns.countplot(x='sex',data=tips)

sns.despine(left=True, bottom=True) #removes top and right spine by default
```

```python
sns.set_style('ticks') #this sets the style for everything now.
```

```python
plt.figure(figsize=(12,3)) #can you figure sizing from matplotlib in seaborn
sns.countplot(x='sex',data=tips)
```

What if we're not presenting for a notebook? What if we're gonna use the graph in a poster?

```python
sns.set_context(context='poster', font_scale=0.8) #context= 'talk', 'paper', 'notebook' <- default
#note this context affects the whole notebook when doing plots

sns.countplot(x='sex',data=tips)
```

```python
sns.set_context(context='notebook', font_scale=1)

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='magma')
```
