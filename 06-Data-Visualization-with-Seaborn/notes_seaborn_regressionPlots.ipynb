```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## L.M. (Linear Model) Plot:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','v'])

#note adding hue gives us 2 scatter plots and 2 linear fits
# Inference: Both male and female have same linear fit in regards to total_bill and tip, so...

#markers come from matplotllib marker symbols.
# we pass in a list of markers cuz we have 2 things (male/female) we change our markers for
```

Since matplotlib is what runs under the hood of seaborn, we can pass directly pass in matplotlib style parameters into seaborn using `scatter_kws={}` and pass in a dictionary. Note this syntax will differ and so reference the documentation.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','v'], 
          scatter_kws={'s':80}) #s affects the size of the markers

# so this scatter_kws is essentially a direct call to matplotlib under the hood, 
#  we are passing in a dictionary of matplotlib arguments
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')

#instead of using hue, which puts both sex on one axes, we can have them on 2 separate columns on the grid. 

#this throws back to the grid lecture, but this is a little easier than dealing with FacetGrid
```

Changing aspect ratio of the grid below:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6, height=8)

#we can change height(size) and aspect ratio
#aspect: For each separate grid. aspect=0.6 means 0.6 width for 1 height. aspect=1.5 means 1.5 width for 1 height.
#height: changes the height of the grid, basically affecting the size of it
```
