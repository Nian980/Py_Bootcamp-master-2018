** What is the max Close price for each bank's stock throughout the time period?**

```python
#xs - crosssection allows us to grab things inside multi-level indexes
temp_df = bank_stocks.xs(key='Close',axis=1,level='Stock Info')
temp_df.head()
#http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.xs.html
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
#note: we are creating a column for the returns DataFrame we initialized above
for tick in tickers:
    returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change() 
#calculates percent change from prev row, hence why first row is NaN
returns.head()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize=(12,6))

bank_stocks['BAC']['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 Day Average', 
                                                                                         ylim=(10,50),
                                                                                         xlim=('2008-01-01','2009-01-01'),
                                                                                         cmap='viridis')

bank_stocks['BAC']['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC Close')

plt.legend()

#Note pandas has a .rolling(window=) method
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rolling.html
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
#build matrix df first:
banks_corr = bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr()
banks_corr
```

```python
sns.heatmap(banks_corr, annot=True, cmap='magma')
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(banks_corr, annot=True, cmap='coolwarm')
```

```python
banks_corr.iplot(kind='heatmap',colorscale='rdylbu')
```
