```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

2 ways to plot something:
### Functional method:

```python
plt.plot(x,y)
#need plt.show() if not in Jupyter Notebook
plt.xlabel("X label")
plt.ylabel("y label")
plt.title("title here")
```

Putting multi-plots on the same canvas

```python
plt.subplot(1,2,1) #rows, columns, plot no. you're referring to
plt.plot(x,y,'r') #color of line is red

plt.subplot(1,2,2)
plt.plot(y,x,'b') #color of line is blue
```

### Object-oriented method: (better way)
Instantiate figure objects, and then call it's methods and attributes.

```python
fig = plt.figure() #create figure object

# imagine a figure object as a blank canvas, and we can add an axis to it:

#we add axes by passing in a list:
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  #left,bottom,width,height   
#the numbers are between 0 and 1, i.e. a percent of the canvas you will take

#now we will plot on that axes:
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Our title')
```

```python
#now we will put 2 sets of figures on one canvas

fig = plt.figure()  #create empty figure object

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('big plot')

axes2.plot(y,x)
axes2.set_title('small plot')
```

### Create subplots using OOP method:

```python
fig,axes = plt.subplots(nrows=3, ncols=3)

plt.tight_layout()  #fixes cramped axes boxes
```

```python
# Use similar to plt.figure() except it uses tuple unpacking to grab fig and axes
fig, axes = plt.subplots(nrows=1, ncols=2)
#subplots is essentially an axes manager on top of plt.figure

# for axe in axes: #we can iterate through axes cuz it's an array!
#     axe.plot(x,y)

axes[0].plot(x,y) #grab first matplotlib.axes and plot on it
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')

plt.tight_layout() #takes care of any overlapping
```

```python
axes #this is just an array of matplotlib axes, which are the ones we manually created when we did fig.add_axes()
#also means we can iterate through them like above!
```

The ONLY difference between doing `plt.subplot(nrows, ncols)` and `plt.figure()` is that subplots is doing the add_axes() thing for you based on the number of rows and columns

## Figure size and DPI:

```python
fig = plt.figure(figsize=(8,2), dpi=100) #figsize = tuple (width, height) in inches

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2), dpi=200) #figsize = tuple (width, height) in inches

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
#How to save the figure.
fig.savefig('my_picture.png') #can also specify dpi here if you want
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X_squared') #all the label is for legend to know which plot to label what
ax.plot(x,x**3,label='X_cbued')

#Add in a legend
ax.legend(loc=0) #loc is location code, check documentation: 0 = 'best', 1='upper right' ...etc. up to 10

#you can also do loc=(0.1,0.1) where it's (x,y) location of bottom left corner of the legend box
```

## Changing plot appearance:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=4,alpha=0.5,ls='-',marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='blue')


#color can also take RGB hex codes, e.g. #FF4B36
#lw = linewidth=1 is default, 2 is thicker...etc.
#alpha controls transparency
#ls = linestyle='-' is default. note: 'steps' also works!
#marker= https://matplotlib.org/api/markers_api.html
#markersize=1 default size of markers
#markerfacecolor='yellow' changes the color of what's inside the marker, good for O-type markers
#markeredgewidth=1 makes the circle line thicker
#markeredgecolor='blue' changes the color of the circle line
```

### Setting x and y limits:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',lw=2,ls='--')

ax.set_xlim([0,1]) #so x-axis goes from 0 to 1, almost like zooming into the line
ax.set_ylim([0,5]) #so y-axis goes from 0 to 5
```
